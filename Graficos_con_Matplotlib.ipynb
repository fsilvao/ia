# Matplotlib

Comprender los elementos que componen cualquier gráfica en Matplotlib es fundamental para entender la lógica de funcionamiento de esta librería y poder así sacarle el mayor provecho.

Veamos entonces estos componentes:

## Figura

Es el "lienzo" sobre el cual dibujaremos nuestra gráfica. En realidad es un objeto tipo
matplotlib_figure.png



A medida que elaboramos la gráfica iremos añadiendo *hijos* a este lienzo (ejes, líneas, puntos, grillas, leyendas, etc.)

## Ejes

Es el objeto (tipo `Axes`) donde realmente dibujaremos las gráficas. Los ejes son hijos de la Figura y a su vez los elementos de la gráfica serán hijos de los ejes.

Podemos tener figuras con un sólo par de ejes:

`Figure`:matplotlib_1_axes.png



O con múltiples ejes:

matplotlib_4_axes.png

Y cada par de ejes de la figura tiene unos **atributos**. Algunos ejemplos de atributos son:

- Los *axis* o ejes individuales (eje horizontal, eje vertical)
- El título (que podemos fijar con el método `set_title()`)
- Las etiquetas de los ejes (que podemos fijar con los métodos `set_xlabel()` y `set_ylabel()`)

## Ejes individuales

Son objetos tipo `axis` que a su vez son hijos tanto de los ejes como de la figura.

También los podemos personalizar a través de sus **atributos**, algunos de los cuales son:

- Las marcas (o *ticks*)
- Las etiquetas de las marcas (o *ticklabels*)
- La ubicación de las marcas
- El tipo y tamaño de letra de las marcas

## Las gráficas

Que son como tal los datos que dibujaremos en los Ejes. Estas gráficas serán hijas tanto de los Ejes como de la Figura misma.

Y podremos tener muchos tipos de gráficas (que veremos en detalle en el curso). Por ejemplo podemos tener una gráfica de dispersión:

matplotlib_grafica_dispersion.png

O por ejemplo múltiples gráficas de línea en un mismo par de Ejes:

matplotlib_graficas_lineas.png

Y dada la flexibilidad de Matplotlib, podemos personalizar nuestras gráficas tanto como queramos añadiendo elementos que faciliten su visualización.

Por ejemplo, podemos añadir leyendas y grillas:
